--------------------------

![VPN_1](images/ssh.png)
![VPN_1](images/ssh2.png)

Then update Ubuntu and install <a href="https://git-scm.com/">git</a> and <a href="https://www.vim.org/">Vim</a>.

    sudo apt update
    sudo apt install git vim -y
    
Now we clone this repo:

    git clone https://github.com/mdh266/JupyterOnGCP.git

Go into it:

    cd JupyterOnGCP
    
Run the script to install <a href="https://docs.conda.io/en/latest/miniconda.html">miniconda</a> with the command:

    source install.sh


I used a similar set up to the official <a href="https://hub.docker.com/r/continuumio/miniconda3">Dockerfile</a> for Miniconda3. I decided against <a href="https://www.docker.com/">Docker</a> as it introduces complexities I did not want to get into.  However, one complexity that arose the way I did things is that you don't have ability to install into a base conda environment so I had to create a miniconda environment and connect it to Jupyter as described <a href="https://ipython.readthedocs.io/en/stable/install/kernel_install.html">here</a>.  If the script ran successfully you will see similar result:


![success](images/success.png)

You can then start the Jupyter notebooks with the command:

    jupyter notebook --ip=0.0.0.0 --port=8888 --no-browser

![success](images/start_notebook1.png)
    
Copy the link replacing 127.0.0.1 to the `external ip` address of your instance:

![success](images/start_notebook2.png)

Then click on Notebook.ipynb

![success](images/start_notebook3.png)
